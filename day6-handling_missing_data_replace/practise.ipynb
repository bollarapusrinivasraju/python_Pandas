```python
import pandas as pd
import numpy as np
df = pd.read_csv('weather_data.csv')
df
```

```python
#replace -99999 with NAN
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
#replace multiple values -99999,-88888 with NAN
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
#replace multiple values based on column names with NAN
new_df = df.replace({
    'temperature':-99999,
    'windspeed' : -99999,
    'event': '0'
    },np.NaN)
new_df
```

```python
#replace the value -99999 to Nan and No event to sunny word

new_df = df.replace({
    -99999:np.NaN,
    'No Event' : 'Sunny'
})
new_df
```

```python
#changing from mph to '' and F to '' based on column using regex
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
#changing from mph to '' and F to '' based on column using regex
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
    },'',regex=True)
new_df
```

```python
#replace a list of value with another list
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
#replace the string to number
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df
```
